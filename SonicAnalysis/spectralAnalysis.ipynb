```python
import numpy as np
```

```python
with open('SpectrumData.txt', 'r') as f:
    lines = f.readlines()
```

```python
spectrumData = []

for line in lines:
    amplitudes = [float(x) for x in line.split()]
    spectrumData.append(np.array(amplitudes))
spectrumData = np.array(spectrumData)
```

```python
maxAmplitudes = np.max(spectrumData, axis=0)
minAmplitudes = np.min(spectrumData, axis=0)
meanAmplitudes = np.mean(spectrumData, axis=0)
stdDevAmplitudes = np.std(spectrumData, axis=0)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(maxAmplitudes.shape[0]), maxAmplitudes)
plt.plot(range(maxAmplitudes.shape[0]), minAmplitudes)
plt.plot(range(maxAmplitudes.shape[0]), meanAmplitudes)
plt.legend(["Max", "Min", "Mean"])
plt.show()
```

```python
normalizingFactor = 1/(maxAmplitudes - minAmplitudes)
```

```python
normalizingFactor[-1] = 1
```

```python
', '.join([f"{x:.4f}f" for x in normalizingFactor.tolist()])
```

```python
normalizedSpectralData = spectrumData / (maxAmplitudes - minAmplitudes)
```

```python
f = normalizedSpectralData[:,0]
s = normalizedSpectralData[:,90]
```

```python
plt.plot(range(len(f)), f)
plt.plot(range(len(s)), s)
```
